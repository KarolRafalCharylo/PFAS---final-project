```python
from pathlib import Path
from src.tracking.run_tracking import run_tracking
import cv2
import numpy as np
import pandas as pd
from src.tracking.position_calculation import project_3d_to_2d_left

root_dir = Path().cwd().parent
data_dir = root_dir / "data"
seq = data_dir / "raw/final_project_2023_rect/seq_03"

#run_tracking(seq)

dp_left = seq / "image_02/data"
glob_left = sorted(dp_left.glob("*.png"))
df = pd.read_csv(data_dir / "test.csv")

result_dir = data_dir / "results"
result_dir.mkdir(parents=True, exist_ok=True)
```

```python
for i, img_left_path in enumerate(glob_left):

    img_left = cv2.imread(str(img_left_path))
    df_frame = df[df["frame"] == i]


    for _, row in df_frame.iterrows():

        x_min_current = row["bbox_cx"]-row["bbox_w"]/2
        y_min_current = row["bbox_cy"]-row["bbox_h"]/2
        x_max_current = row["bbox_cx"]+row["bbox_w"]/2
        y_max_current = row["bbox_cy"]+row["bbox_h"]/2

        pred_ul, pred_vl = project_3d_to_2d_left([row["prediction_x"], row["prediction_y"], row["prediction_z"]])

        x_min_pred = pred_ul-row["bbox_w"]/2
        y_min_pred = pred_vl-row["bbox_h"]/2
        x_max_pred = pred_ul+row["bbox_w"]/2
        y_max_pred = pred_vl+row["bbox_h"]/2

        cv2.rectangle(
            img_left,
            (int(x_min_current), int(y_min_current)),
            (int(x_max_current), int(y_max_current)),
            (0, 255, 0),
            2,
        )

        # cv2.rectangle(
        #     img_left,
        #     (int(x_min_pred), int(y_min_pred)),
        #     (int(x_max_pred), int(y_max_pred)),
        #     (0, 0, 255),
        #     2,
        # )

    img_name = f"{img_left_path.stem}_result.png"
    print(img_name)
    img_path = result_dir / img_name
    cv2.imwrite(str(img_path), img_left)
```
